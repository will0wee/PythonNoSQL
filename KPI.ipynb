```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('resto.csv')

##Les 10 premières ligne du dataset
data.head(10)

##La taille de notre dataset
data.size

#Les colonnes de notre dataset
data.columns

##Récupère la ligne avec l'item le plus commandé
data2 = data.groupby('item_name').quantity.count()
maxi = data2.max()  ##cb de fois l'item a été commandé
item_commander = data2[data2 == maxi].index[0]
print("Item le plus commandé :", item_commander)

##Total d'item commandé
total = 0
for com in data2:
  total += com

print("Total d'item commandé :", total)

##Chiffre d'affaire
CA = 0
for index, row in data.iterrows():
  CA += row['quantity'] * float(row['item_price'][1:])

print("Chiffre d'affaire : ",CA,"$")

#Moyenne de commande :
commandes = []
for index, row in data.iterrows():
  commandes.append(row['quantity'] * float(row['item_price'][1:]))

moy = np.mean(commandes)
print("Moyenne de commande :", moy,'$')
```
